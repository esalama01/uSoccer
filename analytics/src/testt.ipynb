```python
class Product:
    def __init__(self, name, price, stock):
        self.name = str(name)
        self.price = float(price)
        self.stock = int(stock)
        
    def total_value(self):
        return self.price * self.stock

import pandas as pd

df = pd.DataFrame({
    'item': ['Laptop', 'Mouse'],
    'price': [1200, 25],
    'quantity': [5, 50]
})

# Pass entire rows into the class constructor
df.apply(
    lambda row: Product(row['item'], row['price'], row['quantity']), 
    axis=1
)
```

```python
prod_obj = df.iloc[0]
```

```python
prod_obj.total_value()
```
